```python
%matplotlib notebook
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from PIL import Image
from scipy import signal
```

```python
def img_to_arr(img) :
    data = np.asarray(img, dtype="int32")
    return data

img = Image.open("img/robeson-photo.jpg")
img = img_to_arr(img)

iim = np.cumsum(np.cumsum(img,0), 1)

rend = len(iim)
cend = len(iim[0])

b1 = iim[4:,12:]+iim[0:rend-4,:cend-12]-iim[4:,:cend-12]-iim[:rend-4,12:]
bend = len(b1)
f1 = b1[0:(bend-4),:]-b1[4:,:]

f1p = np.pad(f1, (4,6), 'constant')

points = np.where(f1p > 7000)
y = points[0]
x = points[1]


plt.imshow(img, cmap='gray')
plt.scatter(x,y)
plt.show()
```
