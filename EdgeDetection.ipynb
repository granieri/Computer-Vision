### Exploring edge detection in scikit

I used this notebook to experiment with different edge detection algorithms in the scikit library.

Exercise prompt from [here](http://www.cs.smith.edu/~nhowe/teaching/csc370/Assign/edge.html).

```python
%matplotlib inline
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from PIL import Image

from scipy import ndimage as ndi
from skimage.filters import roberts, sobel, scharr, prewitt, laplace
```

```python
img_wrench = Image.open("wrench.jpg")
img_albert = Image.open("albert.jpg")
img_vasily = Image.open("vasily.jpg")
img_valley = Image.open("valley.jpg").convert('LA')
```

```python
roberts_wrench = roberts(img_wrench)
plt.axis('off')
plt.imshow(roberts_wrench, cmap='gray')

sobel_wrench = sobel(img_wrench)
plt.imshow(sobel_wrench, cmap='gray')

prewitt_wrench = prewitt(img_wrench)
plt.imshow(prewitt_wrench, cmap = 'gray')

scharr_wrench = scharr(img_wrench)
plt.imshow(scharr_wrench, cmap = 'gray')

laplace_wrench = laplace(img_wrench)
plt.imshow(laplace_wrench, cmap = 'gray')
plt.show()
```
